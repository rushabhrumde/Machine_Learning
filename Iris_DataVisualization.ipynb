```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.show()
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.describe()
```

```python
np.round(iris.mean(),2)
```

```python
np.round(iris.median(),2)
```

```python
iris.count()
```

```python
pd.crosstab(index=iris["species"],columns='Frequency')
```

```python
coorelation = iris.corr()
coorelation
```

```python
sns.heatmap(coorelation,cmap="RdYlGn",annot=True)
plt.show();
```

```python
sns.heatmap(coorelation,cmap="Blues",annot=True)
plt.show();
```

```python
np.round(iris.quantile((np.arange(0,1,0.05))),2)
```

```python
sns.pairplot(iris, hue="species",palette="RdYlGn",size=5)
plt.show();
```

```python
sns.pairplot(iris,hue="species",palette="copper",size=5,x_vars=["sepal_width","sepal_length"],y_vars=["petal_width","petal_length"])
plt.show()
```

```python
sns.pairplot(iris,kind="reg")
plt.show()
```

```python
sns.pairplot(iris,size=3,vars=["sepal_width","sepal_length"])
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplots_adjust(hspace=0.2,wspace=0.2)
plt.subplot(2,2,1)
plt.title("Sepal Length")
plt.plot(iris.sepal_length,'r')
plt.subplot(2,2,2)
plt.title("Sepal Width")
plt.plot(iris.sepal_length,'b')
plt.subplot(2,2,3)
plt.title("Petal Length")
plt.plot(iris.sepal_length,'g')
plt.subplot(2,2,4)
plt.title("Petal Length")
plt.plot(iris.sepal_length,'y')
plt.show();
```

```python
plt.figure(figsize=(15,10))
sns.distplot(iris.sepal_width)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(data=iris,orient='v',palette="Set2")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x='species',y='sepal_length',data=iris,linewidth=2)
plt.show()
```
